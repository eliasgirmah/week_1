```python
# Correlation Analysis Notebook

# Step 1: Import necessary libraries and modules
import os
import sys
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Adjust the path to include the src directory
sys.path.insert(0, os.path.abspath(os.path.join(os.path.dirname(''), '..', 'src')))

# Import the CorrelationAnalysis class
from correlation_analysis import CorrelationAnalysis

# Step 2: Load the data
base_path = os.path.abspath(os.path.join(os.path.dirname(''), '..', 'data'))
data_path = os.path.join(base_path, 'yfinance_data', 'AAPL_historical_data.csv')

# Check if the file exists
if not os.path.exists(data_path):
    print(f"Data file not found: {data_path}")
else:
    dataframe = pd.read_csv(data_path)
    print("Data Loaded Successfully")

# Display the first few rows of the dataframe
display(dataframe.head())

# Step 3: Initialize CorrelationAnalysis and calculate the correlation matrix
correlation_analysis = CorrelationAnalysis(dataframe)

# Calculate the correlation matrix
correlation_matrix = correlation_analysis.calculate_correlation()
print("Correlation Matrix:")
display(correlation_matrix)

# Step 4: Plot the correlation heatmap
correlation_analysis.plot_correlation_heatmap(correlation_matrix)
```
